```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

corus = ["natural language process is fun.", "Text preprocessing is an important NLP step."]
```

```python
bow = CountVectorizer()
bow_matrix = bow.fit_transform(corus)
print("Bow Feature Name : ", bow.get_feature_names_out())
print("Bow Matrix:\n  ", bow_matrix.toarray())
```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(corus)
print("TF-IDF Feature Name : ", tfidf.get_feature_names_out())
print("TF-IDF Matrix : \n", tfidf_matrix.toarray())
```
