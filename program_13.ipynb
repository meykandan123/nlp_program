```python
import heapq

# Graph representation (Adjacency List)
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

# Heuristic values (estimated cost to goal F)
heuristic = {
    'A': 5,
    'B': 4,
    'C': 2,
    'D': 3,
    'E': 1,
    'F': 0
}

def best_first_search(start, goal):
    visited = set()
    priority_queue = []
    
    # Push start node with its heuristic value
    heapq.heappush(priority_queue, (heuristic[start], start))
    
    path = []

    while priority_queue:
        h, node = heapq.heappop(priority_queue)

        if node not in visited:
            print("Visiting:", node)
            visited.add(node)
            path.append(node)

            if node == goal:
                print("Goal reached!")
                return path

            # Add neighbors to priority queue
            for neighbor in graph[node]:
                if neighbor not in visited:
                    heapq.heappush(priority_queue, (heuristic[neighbor], neighbor))

    return path


# Run the algorithm
result = best_first_search('A', 'F')
print("Path followed:", result)
```
